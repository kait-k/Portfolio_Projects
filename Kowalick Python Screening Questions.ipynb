<b> Q2. Please re-name all the newly filtered columns in the pandas dataframe to the following: </b>

Employment Type, Degree Status, Marriage-Status, Job-Role, Family-Role, Ethnicity, Gender, Country, Earnings

E.g. State-Gov becomes Employment Type, Bachelors becomes Degree Status etc.

<b> Put your code below </b>

```python
#I created a dictionary of the name modifications, then a function to iterate over the dictionary and rename the columns inplace
column_rename  = {' State-gov':'Employment Type', ' Bachelors' : 'Degree Status', ' Never-married' : 'Marriage-Status',
' Adm-clerical' : 'Job-Role', ' Not-in-family' : 'Family-Role', ' White' : 'Ethnicity', ' Male' : 'Gender', 
' United-States' : 'Country',' <=50K' : 'Earnings'}

def rename_columns(df, column_rename):
    df.rename(columns = column_rename, inplace = True)

rename_columns(df_filtered, column_rename)
df_filtered = df_filtered.drop(df_filtered[df_filtered['Job-Role'] == ' ?'].index)
df_filtered
```

<b> Q3. The Job Role Columns holds the job information for each individual in this census snapshot. Using this column, please proceed to create a Bar Chart that shows the count of 'Unique' Jobs per Job Group in the "Job-Role" Column in ascending order as per the provided image below </b>


<b> Put your code below </b>

```python
import seaborn as sns
#I created the plot using seaborn then created a variable to allow the plot to be ordered by ascending order. .index required to just extract
#the value names vs the name and the value
order = df_filtered['Job-Role'].value_counts().sort_values(ascending=True).index
sns.countplot(x = 'Job-Role', data = df_filtered, order = order)
plt.title('Unique Jobs per Job Group')
plt.xticks(rotation=90)
plt.show()
```

<b> Q4. Please create two bar plots as per below that show:
    
    1) The number of individuals who have a High School Graduate Diploma AND earn <=50K in the United States
    2) The number of individuals who have a High School Graduate Diploma AND earn >50K in the United States

```python
#to ensure I was using the correct variables 
print(df_filtered.head())
print(df_filtered['Degree Status'].unique())
print(df_filtered['Earnings'].unique())
print(df_filtered['Country'].unique())
```

```python
under_50k = df_filtered[(df_filtered['Degree Status'] == ' HS-grad') & (df_filtered['Earnings'] ==' <=50K') & (df_filtered['Country'] ==' United-States')]
order = under_50k['Job-Role'].value_counts().sort_values(ascending=False).index
sns.countplot(x = 'Job-Role', data = under_50k, order = order)
plt.title('Individuals in the USA with HS Diploma and Salary <=50k')
plt.xticks(rotation=90)
plt.xlabel('Job-Role')
plt.ylabel('Count')
plt.show()

over_50k = df_filtered[(df_filtered['Degree Status'] == ' HS-grad') & (df_filtered['Earnings'] ==' >50K') & (df_filtered['Country'] ==' United-States')]
order2 = over_50k['Job-Role'].value_counts().sort_values(ascending=False).index

sns.countplot(x= 'Job-Role', data = over_50k, order = order2)
plt.title('Individuals in the USA with HS Diploma and Salary >50k')
plt.xticks(rotation=90)
plt.xlabel('Job-Role')
plt.ylabel('Count')
plt.show()
```

<b> Q5. Which Job Role has the highest <i> proportion </i> of individuals who earn >50K? </b>

<b> Put your code below </b>

```python
#not filtered by country or degree
#>50k/total count * 100 
over_50k_filter = df_filtered[df_filtered['Earnings'] == ' >50K'] #filter
over_50k_count = over_50k_filter['Job-Role'].value_counts() #>50k job count
total_count = df_filtered['Job-Role'].value_counts() #total job count
prop_per_job = round(((over_50k_count/ total_count) * 100),2).sort_values(ascending = False) #math
print(prop_per_job)
```
